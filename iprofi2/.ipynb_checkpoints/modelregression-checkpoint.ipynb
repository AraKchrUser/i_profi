```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
def convert(s):
    if type(s) is float or type(s) is int:
        return s
    if all([c.isalpha() and c != '-' for c in s]):
        return float(s)
    else:
        if ',' in s:
            c = ','
            s = s.replace(s[:s.index(c)+1], '0.')
        elif '-' in s:
            c = '-'
            s = s.replace(s[:s.index(c)+1], '0.')
        
        return float(s)
```

```python
trn = pd.read_csv('train.csv', encoding='cp1251')
y = trn['critical_temperature'].values
X = trn.drop('critical_temperature', axis=1).values

Xt = pd.read_csv('test.csv', encoding='cp1251').values # test

# news : add formulae
form = pd.read_csv('formula_train.csv', encoding='cp1251')
form_tst = pd.read_csv('formula_test.csv', encoding='cp1251')
for col in form.loc[:, form.columns[:-1]].columns:
    if form[col].dtype == 'object':
        form[col] = form[col].apply(convert)

#X = pd.concat((pd.DataFrame(X), form.loc[:, form.columns[:-1]]), axis=1).values
#Xt = pd.concat((pd.DataFrame(Xt), form.loc[:, form_tst.columns[:-1]]), axis=1).values

#X = pd.concat((pd.DataFrame(X), form.loc[:, form.columns[:-1]].sum(axis=1)), axis=1).values
#Xt = pd.concat((pd.DataFrame(Xt), form.loc[:, form_tst.columns[:-1]].sum(axis=1)), axis=1).values

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42) #X, Xt, y, 0 # 

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
#from sklearn.preprocessing import Normalizer
#scaler = Normalizer()
#scaler.fit(X_train)
#X_train = scaler.transform(X_train)
#X_test = scaler.transform(X_test)
#data = pd.DataFrame(X_train)
#data['target'] = pd.Series(y)

from sklearn.feature_selection import SelectKBest, f_regression
select = SelectKBest(f_regression, k=20)
select.fit(X_train, y_train)
# выделить и для тренировочного
train = select.transform(X_train)
test = select.transform(X_test)
```

```python
from keras import backend as K

def coeff_determination(y_true, y_pred):
    from keras import backend as K
    SS_res =  K.sum(K.square( y_true-y_pred ))
    SS_tot = K.sum(K.square( y_true - K.mean(y_true) ) )
    return ( 1 - SS_res/(SS_tot + K.epsilon()) )

def build_model():
    #model = keras.Sequential([
    #    layers.Dense(64, activation='relu', input_shape=[len(pd.DataFrame(train).keys())]),
    #    layers.Dense(64, activation='relu'),
    #    layers.Dense(1)
    #])
    model = keras.Sequential([
        layers.LSTM(32, input_shape=(None, train.shape[-1]), return_sequences=False),
        layers.Dense(1)
        ])
    optimizer = tf.keras.optimizers.RMSprop(0.001)
    model.compile(loss='mse', optimizer=optimizer, metrics=[coeff_determination])
    return model
```

```python
model = build_model()
model.summary()
```

```python
import numpy as np
class PrintDot(keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs):
        if epoch % 100 == 0: print('')
        print('.', end='')
EPOCH = 1000
early_stop = keras.callbacks.EarlyStopping(monitor='val_loss', patience=10)
%time
tr = np.expand_dims(train, 1)
history = model.fit(tr, y_train, epochs=EPOCH, validation_split=0.2, verbose=0, callbacks=[early_stop, PrintDot()])
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, model.predict(np.expand_dims(test, 1)))
```
