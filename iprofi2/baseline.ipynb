# отбор признаков

```python
from sklearn.feature_selection import SelectKBest, f_regression
select = SelectKBest(f_regression, k=20)
select.fit(X_train, y_train)
# выделить и для тренировочного
test = select.transform(X_train)
```

```python
df = pd.concat([pd.DataFrame(test), pd.DataFrame(y_train)], axis=1)
```

```python
fig, ax = plt.subplots(figsize=(15,15)) 
sns.heatmap(df.corr(), cmap='coolwarm', annot=True)
```

# модели

```python
from sklearn.metrics import r2_score
from sklearn.linear_model import LassoCV, RidgeCV
```

```python
reg = LassoCV(cv=5, random_state=0).fit(test, y_train)
```

```python
reg.score(test, y_train)
```

```python
#r2_score(y_test, reg.predict(select.transform(X_test)))
reg.predict(select.transform(X_test))
```

```python
ser = pd.Series(reg.predict(select.transform(X_test)))
ser.to_csv('answer.csv.csv', index=False)
```

```python
clf = RidgeCV(alphas=np.logspace(-6, 6, 13), cv=7).fit(test, y_train)
```

```python
clf.score(test, y_train)
```

```python
clf.predict(select.transform(X_test))
```

```python
ser = pd.Series(clf.predict(select.transform(X_test)))
ser.to_csv('answer.csv', index=False)
```

```python
r2_score(y_test, clf.predict(select.transform(X_test)))
```
